```python
#This is a script of forward difference method of the advection-diffusion equation

import numpy as np
import matplotlib.pyplot as plt

#set parameters
DT = 0.05
DX = 1
por = 0.3
Cf = 100.0
Cb = 0.0
Ci = 0.0
DH = 2.5
q = 1.5

L = 50
Tmax = 5
N = int(L / DX) + 1

node_locations = np.arange(0, L + DX, DX)
#print(node_locations)
number_timesteps = int(Tmax / DT)
#print(number_timesteps)
time_increments = np.arange(DT, Tmax + DT, DT)
#print(time_increments)
Cold = np.zeros(N) * Ci
Cnew = np.zeros(N)
#print(Cold)
Cold[0] = Cf
#print(Cnew)
Cnew=Cold
#print(Cnew)



#set coefficients in in FD equations
p = (DH * DT) / (por * DX**2)
r = (q * DT) / (2 * por * DX)

a = p - r
b = 1.0 - 2 * p
c = p + r


NPRINT = 0

itime = 1
for i in np.arange(DT, Tmax + DT, DT):
    Cnew[0] = Cf
    Cnew[50] = Cb
    NPRINT += 1
    itime += 1
       
    for k in np.arange(1,N-2,DX):
        Cnew[k] = a * Cold[k + 1] + b * Cold[k] + c * Cold[k - 1]

#print(Cnew)

    
```
